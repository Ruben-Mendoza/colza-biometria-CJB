```python
def obj_to_num(dataframe, column_name):
    # Recorrer todos los valores de la columna y reemplazar las comas por puntos
    dataframe[column_name] = dataframe[column_name].apply(lambda x: x.replace(',', '.') if isinstance(x, str) else x
                                                    ).astype(float)
```

```python
columns = ['α Tocopherol', 'γ Tocopherol', 'δ Tocopherol', 'α+γ+δ Tocopherol',
           'Palmitic Acid', 'Palmitoleic Acid', 'Stearic Acid', 'Oleic Acid',
           'Linoleic Acid', 'Linolenic Acid', 'Arachidic Acid', 'Ecosenoic Acid',
           'OSI', 'Oil content', 'IV'
           ]

for col in columns:
    obj_to_num(df_1, col)
```

```python
def imputar_columna_faltante(df, columna_suma, columna1, columna2, columna_faltante):
    valores_faltantes = df[columna_faltante].isnull()
    df.loc[valores_faltantes, columna_faltante] = df.loc[valores_faltantes, columna_suma] - df.loc[valores_faltantes, columna1] - df.loc[valores_faltantes, columna2]
```

```python
def plot_numeric_distribution(dataframe, column_name):
    plt.figure(figsize=(10, 6))
    dataframe[column_name].plot(kind='density')
    plt.xlabel(column_name)
    plt.ylabel('Frecuencia')
    plt.title(f'Distribución de valores de la columna "{column_name}"')
    plt.grid(True)
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer
```

```python
escalador_1 = MinMaxScaler()
df_1_scaled_OSI = escalador_1.fit_transform(df_1_encoded_OSI)

for k in range(1,7):

    imputer = KNNImputer(n_neighbors = k, weights = 'distance')
    df_1_imputed = imputer.fit_transform(df_1_scaled_OSI)

    df_1_inv_scaled = escalador_1.inverse_transform(df_1_imputed)
    df_1_clean = pd.DataFrame(df_1_inv_scaled, columns = df_1_encoded_OSI.columns)

    plt.figure(figsize=(10, 6))
    df_1['OSI'].plot(kind='density', label = 'Sin imputar')
    df_1_clean['OSI'].plot(kind='density', label = 'imputado')

    imp_1 = df_1_clean.iloc[43]['OSI']
    vec_1 = df_1.iloc[42]['OSI']

    imp_2 = df_1_clean.iloc[69]['OSI']
    vec_2 = df_1.iloc[68]['OSI']

    print(f'Valor Imputado : {imp_1}')
    print(f'Valor Vecino {vec_1}')

    print(f'Valor Imputado : {imp_2}')
    print(f'Valor Vecino {vec_2}')

    plt.xlabel('OSI')
    plt.ylabel('Frecuencia')
    plt.title(f'Distribución de valores de la columna "OSI" con k = {k}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
df_1_encoded = pd.get_dummies(data = df_1, columns = ['Localidad', 'Cultivar'], dtype = int)
df_1_encoded_Oil = df_1_encoded.drop(['Código'], axis = 1)

escalador_2 = MinMaxScaler()
df_1_scaled_Oil = escalador_2.fit_transform(df_1_encoded_Oil)

dist = 0

for k in range(1,7):

    imputer = KNNImputer(n_neighbors = k, weights = 'distance')
    df_1_imputed = imputer.fit_transform(df_1_scaled_Oil)

    df_1_inv_scaled = escalador_2.inverse_transform(df_1_imputed)
    df_1_clean = pd.DataFrame(df_1_inv_scaled, columns = df_1_encoded_Oil.columns)

    plt.figure(figsize=(10, 6))
    df_1['Oil content'].plot(kind='density', label = 'Sin imputar')
    df_1_clean['Oil content'].plot(kind='density', label = 'imputado')

    imp_1 = df_1_clean.iloc[35]['Oil content']
    vec_1 = df_1.iloc[34]['Oil content']
    sum1 = abs(imp_1 - vec_1)

    imp_2 = df_1_clean.iloc[43]['Oil content']
    vec_2 = df_1.iloc[42]['Oil content']
    sum2 = abs(imp_2 - vec_2)

    imp_3 = df_1_clean.iloc[57]['Oil content']
    vec_3 = df_1.iloc[56]['Oil content']
    sum3 = abs(imp_3 - vec_3)

    imp_4 = df_1_clean.iloc[58]['Oil content']
    vec_4 = df_1.iloc[59]['Oil content']
    sum4 = abs(imp_4 - vec_4)

    imp_5 = df_1_clean.iloc[69]['Oil content']
    vec_5 = df_1.iloc[68]['Oil content']
    sum5 = abs(imp_5 - vec_5)

    dist = (sum1 + sum2 + sum3 + sum4 + sum5) / 5

    print(f'Valor Imputado : {imp_1}')
    print(f'Valor Vecino {vec_1}')

    print(f'Valor Imputado : {imp_2}')
    print(f'Valor Vecino {vec_2}\n(iguales)')

    print(f'Valor Imputado : {imp_3}')
    print(f'Valor Vecino {vec_3}')

    print(f'Valor Imputado : {imp_4}')
    print(f'Valor Vecino {vec_4}')

    print(f'Valor Imputado : {imp_5}')
    print(f'Valor Vecino {vec_5}\n(iguales)\n')

    print(f'La distancia promedio entre vecinos es : {dist}')

    plt.xlabel('Oil Content')
    plt.ylabel('Frecuencia')
    plt.title(f'Distribución de valores de la columna "Oil content" con k = {k}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def obj_to_date(dataframe, column_name):
    # Función para convertir valores con la forma dd/mm/yyyy a datetime
    def convert_to_datetime(date_str):
        try:
            return pd.to_datetime(date_str, format='%d/%m/%Y')
        except:
            return date_str

    # Recorrer todos los valores de la columna y convertir a datetime si es una cadena con la forma dd/mm/yyyy
    dataframe[column_name] = dataframe[column_name].apply(lambda x: convert_to_datetime(x) if isinstance(x, str) else x)
```
